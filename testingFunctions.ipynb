```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import Axes #type: ignore
import numpy as np
from numpy._typing import _16Bit
# Asegurarse de que las probabilidades sumen 1
probTerreno = [0.95, 0.05]
probTerreno /= np.array(probTerreno).sum()

def GenerarMapa(tamano = 20,dispercion = 1,segmentos = 6):

	sizeMapa = (tamano,tamano)

	centro = tamano // 2

	arrayTerreno = np.zeros(sizeMapa)
	for y in range(tamano):
		for x in range(tamano):
			distancia = np.sqrt((y - centro)**2 + (x - centro)**2)*1.8
			valor = segmentos * (np.random.randint(tamano-dispercion,tamano) - distancia) / tamano
			arrayTerreno[y, x] = max(0, valor)  # Asegurarse de que los valores sean no negativos

	arrayTerreno = np.int16(arrayTerreno)
	print(arrayTerreno)
	return arrayTerreno

reglaNiveles = [
    (1,3),
	(4,9),
	(10,19),
	(20,29),
	(30,39),
	(40,49),
	(50,59),
	(60,69),
	(70,79),
	(80,89),
	(90,99),
	(100,109),
	(110,120),
    ]

def GenerarNivelDificultad(reglaNiveles = [(1,9),(10,24),(25,34),(35,54),(55,69),(70,80)], mapa = np.zeros((4,4))):
	if reglaNiveles and isinstance(reglaNiveles,list): pass
	else : raise Exception('vacio o invalido', 'No hay dato de regla de niveles')

	# if mapa : pass
	# else : raise Exception('vacio o invalido', 'Esta vacio o no hay datos del mapa(numpy)')
    
	sizeMap = len(mapa)
	arrayNivelDificultad = np.zeros((sizeMap,sizeMap))
	for y in range(sizeMap):
		for x in range(sizeMap):
			lvl_min, lvl_max = reglaNiveles[mapa[y,x]]
			arrayNivelDificultad[y,x] = np.random.randint(lvl_min,lvl_max)

	arrayNivelDificultad = np.int16(arrayNivelDificultad)
	print(arrayNivelDificultad)
	return arrayNivelDificultad

mapa = GenerarMapa(tamano=30, dispercion=3,segmentos=3)

mapaNiveles = GenerarNivelDificultad(reglaNiveles=reglaNiveles,mapa=np.array(mapa))

ax : Axes
fig, ax = plt.subplots()
fig.set_size_inches((30,30))
ax.matshow(mapa,cmap='terrain') #type: ignore

mapa = np.array(mapa)
for y in range(len(mapa)):
	for x in range(len(mapa)):
		ax.text(y,x,str(mapaNiveles[y,x]), va='center', ha='center') #type: ignore

print(ax)
```
